# Ploting Histograms

```python
import numpy as np
ml_students_age = np.random.randint(18,45,(100))
py_students_age = np.random.randint(15,45,(100))
```

```python
plt.hist(ml_students_age)
plt.title("Histogram",fontsize=25)
plt.xlabel("X-axis",fontsize=15)
plt.ylabel("Y-axis",fontsize=15)
plt.show()
```

```python
bins = [15,20,25,30,35,40,45] #depand on x-axis
plt.hist(ml_students_age,bins, rwidth=0.8)
plt.title("Histogram",fontsize=25)
plt.xlabel("X-axis",fontsize=15)
plt.ylabel("Y-axis",fontsize=15)
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins, rwidth=0.8,histtype='step')
plt.title("Histogram",fontsize=25)
plt.xlabel("X-axis",fontsize=15)
plt.ylabel("Y-axis",fontsize=15)
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins, rwidth=0.9,histtype='bar')
plt.hist(py_students_age,bins, rwidth=0.8,histtype='bar')
plt.title("Histogram",fontsize=25)
plt.xlabel("X-axis",fontsize=15)
plt.ylabel("Y-axis",fontsize=15)
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist([ml_students_age,py_students_age],bins, rwidth=0.9, color=['r','g'],histtype='bar')
# plt.hist(py_students_age,bins, rwidth=0.8,histtype='bar')
plt.title("Histogram",fontsize=25)
plt.xlabel("X-axis",fontsize=15)
plt.ylabel("Y-axis",fontsize=15)
plt.show()
```

# Scatter

```python
df_google_play_store_app = pd.read_csv("CSV_Files//googleplaystore.csv")
df_google_play_store_app.shape
```

```python
df_google_play_store_app = pd.read_csv("CSV_Files//googleplaystore.csv", nrows=1000)
df_google_play_store_app.shape
```

```python
x = df_google_play_store_app["Rating"]
y = df_google_play_store_app["Reviews"]
```

```python
plt.scatter(x,y, c = "g")
plt.title("Play store data",fontsize = 25)
plt.xlabel("Rating",fontsize = 15)
plt.ylabel("Reviews",fontsize = 15)
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y, c = "g", marker = "o", linewidth = 15, alpha = 0.8, s=100, edgecolor="r")
plt.title("Play store data",fontsize = 25)
plt.xlabel("Rating",fontsize = 15)
plt.ylabel("Reviews",fontsize = 15)
```
