# Series in Pandas

```python
import pandas as pd
```

```python
lit = [1,2,3,"Azeem"]
```

```python
ser = pd.Series(lit)
print(ser)
```

```python
ser[0]
```

```python
ser[2]
```

```python
ser[0:3]
```

```python
type(ser)
```

```python
ser1 = pd.Series(lit)
print(ser1)
```

```python
ser + ser1
```

```python
ser2 = pd.Series(lit, index = ['a','b','c','d'])
print(ser2)
```

```python
ser + ser2
```

```python
ser3 = pd.Series(['1','2','3','4'], index = ['a','b','c','d'], dtype = float)
print(ser3)
```

```python
dect_s = pd.Series({'a':1,'b':2})
dect_s
```

```python
max(ser3)
```

```python
min(ser3)
```
